```python
# Defining column names for a new dataframe
columns = ['id','title','overview','popularity','vote_average','profits']
df = pd.DataFrame(columns=columns)
```

```python
# Making a call to the API for each movie to know the budget and the profits
for film in profits:
    print("Title:",film['title'],
          "\nid:",film['id'],
          "\nOverview:",film['overview'],
          "\nPopularity:",film['popularity'],
          "\nVote average",film['vote_average'],"\n")
    
    film_profits = requests.get('https://api.themoviedb.org/3/movie/'+ str(film['id']) +'?api_key=c2cfd454b2992c642fedcaad890eedce&language=en-US')
    film_profits = film_profits.json()
    # Storing title and profits in my dataframe (profits is called revenue)
    df.loc[len(df)]=[
        film['id'],
        film['title'],
        film['overview'],
        film['popularity'],
        film['vote_average'],
        film_profits['revenue']
    ]
```

```python
df_ordenado = df.sort_values('profits',ascending=False)
df_profits = df_ordenado.drop_duplicates()
```
