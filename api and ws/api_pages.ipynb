```python
page = 1
titles = []
while page != 1472:
      url = f"https://subslikescript.com/movies?page={page}"
      response = requests.get(url)
      html = response.content
      soup = BeautifulSoup(html, "lxml")
      for h3 in soup.find_all("article",{'class':'main-article'}):
            titles.append(h3.get_text(strip=True))
      page = page + 1
```
